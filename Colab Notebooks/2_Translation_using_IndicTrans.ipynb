# Translating Telugu Text to English

```python
import pandas as pd
from indicTrans.inference.engine import Model
from google.colab import files

# Upload the dataset file
uploaded = files.upload()

# Get the file name of the uploaded dataset
uploaded_file_name = list(uploaded.keys())[0]

# Load the uploaded dataset
data = pd.read_excel(uploaded_file_name)
data = data.head(1000)

# Initialize the IndicTrans model
indic2en_model = Model(expdir='../indic-en')

# Function to transliterate Telugu text to English
def transliterate_to_english(text):
    english_text = indic2en_model.translate_paragraph(text, 'hi', 'en')
    return english_text

# Apply the transliteration function to the "telugu_text" column and store the result in a new column
data['english_text'] = data['preprocessed_text'].apply(transliterate_to_english)

# Save the updated dataset with English transliterations
data.to_excel('constraint_hin_eng4to5.xlsx', index=False)
```
